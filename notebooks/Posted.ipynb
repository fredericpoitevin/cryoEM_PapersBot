```python
import numpy as np
```

```python
posted = np.genfromtxt('../posted.dat', str)
```

```python
for post in posted:
    print(post)
```

```python
journal = []
for post in posted:
    journal.append(post.split('/')[2])
```

```python
from collections import Counter

journals  = Counter(journal).keys()
article_published = Counter(journal).values()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(6,6))
plt.xticks(rotation=90)
plt.bar(journals, article_published, width=1, color='black')
plt.grid()
plt.show()
```
